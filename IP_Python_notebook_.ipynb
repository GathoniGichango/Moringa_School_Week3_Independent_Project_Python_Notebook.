```python
# drop unnecessary columns
#

df_combine1 = df_combine.drop(columns = ['COUNTRY_A', 'COUNTRY_B'])
df_combine1
```

```python
# check and remove duplicates
#
df_combine1.duplicated()

# Remove duplicates
#
df_combine1.drop_duplicates()
```
